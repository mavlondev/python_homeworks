```python
def simple_generation():
    print("Hello, World!")
    yield 1
    yield 2
    yield 3

val = simple_generation()
print(next(val))
```

```python
def rrange(start=0, stop=None, step=1):
    if stop is None:
        stop = start
        start = 0
    while start < stop:
        yield start
        start += step

for i in rrange(5):
    print(i)
```

```python
def deco(func):
    def wrapper(*args, **kwargs):
        print("Before function call")
        result = func(*args, **kwargs)
        print("After function call")
        return result
    return wrapper

@deco
def say_hello(name):
    print(f"Hello, {name}!")
say_hello("Alice")
```

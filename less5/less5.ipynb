```python
keys = ["a", "b", "c", "d", "e"]
values = [1, 2, 3, 4, 5]

my_dict = dict(zip(keys, values))
print(my_dict)
```

```python
new_dict = { k:v for (k, v) in zip(keys, values) }
new_dict
```

```python
{ k:v**2 for (k, v) in zip(keys, values) }
```

```python
def invest(amount, rate, years):
    for year in range(1, years + 1):
        amount += amount * rate
        print(f"year {year}: ${amount:.2f}")

invest(100, 0.05, 4)
```

Add new Person class

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"
    
    def eat(self, food):
        return f"{self.name} is eating {food}"
    
person1 = Person("Alice", 30)
print(person1)
print(person1.eat("apple"))
```

Create new Teacher class

```python
class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject
    def __str__(self):
        return f"Teacher(name={self.name}, age={self.age}, subject={self.subject})"
    
    def eat(self, food):
        print(super().eat(food))
        return f"{self.name} is eating {food} while teaching {self.subject}"
    
    
teacher1 = Teacher("Bob", 40, "Math")
print(teacher1)
print(teacher1.eat("sandwich"))
```

```python
class NumberList(list):
    def __init__(self, *args):
        for i in args:
            if not isinstance(i, (int, float)):
                raise ValueError("All elements must be numbers")
        super().__init__(args)
    def append(self, item):
        if not isinstance(item, (int, float)):
            raise ValueError("Only numbers can be added to the list")
        self.append(item)
number_list = NumberList(1, 2, 3)
print(number_list)
```

```python
class AgarageList(list):
    def __init__(self, *args):
        for i in args:
            if not isinstance(i, (int, float)):
                raise ValueError("All elements must be numbers")
        super().__init__(args)
    @property
    def avarage(self):
        return sum(self) / len(self) if self else 0
    

garage_list = AgarageList(1, 2, 3, 4, 5)
print(garage_list.avarage)
```
